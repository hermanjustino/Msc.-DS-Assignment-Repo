# Handle Missing Values
Identify and handle missing values in the dataset, using methods such as filling or dropping them.

```python
# Handle Missing Values

# Check for missing values in the dataset
missing_values = data.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Visualize missing values using a heatmap
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

# Fill missing values with the mean of the column
data_filled = data.fillna(data.mean())

# Verify that there are no missing values left
missing_values_filled = data_filled.isnull().sum()
print("Missing values after filling:\n", missing_values_filled)

# Alternatively, drop rows with missing values
data_dropped = data.dropna()

# Verify that there are no missing values left
missing_values_dropped = data_dropped.isnull().sum()
print("Missing values after dropping rows:\n", missing_values_dropped)
```

```python
pd.set_option('display.max_columns', None)
selected_data = data.iloc[1980:1997]  # iloc is zero-indexed, so row 1980 is at index 1979

# Extract the total population row (row 1980)
total_population = selected_data.iloc[0, 1:]  # Assuming the first column is not a neighborhood

# Calculate the proportion for each neighborhood for rows below the total population row
proportion_data = selected_data.iloc[1:, 1:].div(total_population, axis=1) * 100

# Add the first column back (assuming it contains the labels for educational attainment levels)
proportion_data.insert(0, selected_data.columns[0], selected_data.iloc[1:, 0])

# Combine the total population row with the proportion data
education_by_neighborhood = pd.concat([selected_data.iloc[[0]], proportion_data])

# Display the result data
education_by_neighborhood
```
